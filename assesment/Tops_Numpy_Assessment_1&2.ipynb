## Q 3: Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for a recession period with a non- recession period.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
auto_sales = pd.read_csv("/content/historical_automobile_sales.csv")

# Map 'Recession' column to 'Period'
auto_sales['Period'] = auto_sales['Recession'].map({1: 'Recession', 0: 'Non-Recession'})

# Group by Year, Vehicle_Type, and Period
yearly_sales = auto_sales.groupby(['Year', 'Vehicle_Type', 'Period'])['Automobile_Sales'].sum().reset_index()

# Create subplots for better layout
plt.figure(figsize=(14, 6))

# Recession Plot
plt.subplot(2, 1, 1)
sns.lineplot(
    data=yearly_sales[yearly_sales['Period'] == 'Recession'],
    x='Year',
    y='Automobile_Sales',
    hue='Vehicle_Type',
    marker='o'
)
plt.title('Sales Trend During Recession')
plt.xlabel('Year')
plt.ylabel('Total Automobile Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)

# Non-Recession Plot
plt.subplot(2, 1, 2)
sns.lineplot(
    data=yearly_sales[yearly_sales['Period'] == 'Non-Recession'],
    x='Year',
    y='Automobile_Sales',
    hue='Vehicle_Type',
    marker='o'
)
plt.title('Sales Trend During Non-Recession')
plt.xlabel('Year')
plt.ylabel('Total Automobile Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

## Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
auto_sales = pd.read_csv("/content/historical_automobile_sales.csv")

# Map 'Recession' column to 'Period'
auto_sales['Period'] = auto_sales['Recession'].map({1: 'Recession', 0: 'Non-Recession'})

# Group by Vehicle_Type and Period to get total sales
total_sales = auto_sales.groupby(['Vehicle_Type', 'Period'])['Automobile_Sales'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=total_sales,
    x='Vehicle_Type',
    y='Automobile_Sales',
    hue='Period',
    palette='viridis'
)

# Add labels and title
plt.title('Total Sales of Vehicle Types: Recession vs Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Total Automobile Sales')
plt.legend(title='Period')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
